```python
# Split dataset into Legendary and Regular Pokémon
legendary_pokemon = data[data["category"] == "legendary"] # Change df to data
regular_pokemon = data[data["category"] == "regular"] # Change df to data

# Compute mean stats for both categories
legendary_stats = legendary_pokemon.iloc[:, 7:].mean()  # Stats from 'total_base_stats' onward
regular_stats = regular_pokemon.iloc[:, 7:].mean()

# Combine into a single DataFrame for comparison
comparison_df = pd.DataFrame({"Legendary": legendary_stats, "Regular": regular_stats})

# Plot comparison
plt.figure(figsize=(10, 5))
comparison_df.plot(kind='bar', figsize=(12, 6), colormap='coolwarm', edgecolor='black')
plt.title("Comparison of Legendary vs. Regular Pokémon Stats")
plt.ylabel("Average Stat Value")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Category")
plt.show()

# Display comparison DataFrame
print(comparison_df)
```

```python
import matplotlib.pyplot as plt
# Group data by Generation and count the number of Pokémon in each
generation_counts = data["generation"].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(6, 6))
generation_counts.plot(kind="bar", color="skyblue")

# Labels and title
plt.title("Number of Pokémon by Generation")
plt.xlabel("Generation")
plt.ylabel("Count of Pokémon")
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()

# Count the number of Pokémon per primary type
primary_type_counts = data["primary_type"].value_counts()
plt.figure(figsize=(8, 6))
primary_type_counts.plot(kind="bar", color="red")
plt.title("Number of Pokémon by Primary Type")
plt.xlabel("Primary Type")
plt.ylabel("Count of Pokémon")
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()

# Count the number of Pokémon per category
category_counts = data["category"].value_counts()

# Plot the pie chart
plt.figure(figsize=(10, 6))
plt.pie(
category_counts,
labels=category_counts.index,
autopct="%1.1f%%",
startangle=140,
colors=plt.cm.Paired.colors
)

# Add a title with the total count of Pokémon
plt.title(f"Pokémon Distribution by Category (Total: {data.shape[0]})")

# Show plot
plt.show()
```

```python
# Extract first 5 rows of required columns
import numpy as np # Import numpy and alias it as np
subset = data.iloc[:5][['total_base_stats', 'hp', 'attack']] # Change df to data
labels = data.iloc[:5]['name']


# Create the bar width
x = np.arange(len(labels))  # Label locations
width = 0.3  # Width of the bars

# Create the bar chart
fig, ax = plt.subplots()
rects1 = ax.bar(x - width, subset['total_base_stats'], width, label='Total Base Stats')
rects2 = ax.bar(x, subset['hp'], width, label='HP')
rects3 = ax.bar(x + width, subset['attack'], width, label='Attack')

# Add labels, title, and legend
ax.set_xlabel('Pokemon')
ax.set_ylabel('Stat Value')
ax.set_title('Comparison of Base Stats, HP, and Attack for First 5 Pokemon')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns

# Extract relevant numerical columns for heatmap (last 5 rows)
heatmap_data = data.iloc[-5:, 7:]  # Selecting numerical columns from last 5 rows, Changed df to data
pokemon_names = data.iloc[-5:, 1]


# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Add title
plt.title('Correlation Heatmap of Last 5 Pokemon Stats')

# Show Pokemon names on y-axis
plt.yticks(ticks=np.arange(len(pokemon_names)) + 0.5, labels=pokemon_names, rotation=0)

# Show the plot
plt.show()
```

```python
# Select first 10 Pokémon
df_sample = data.iloc[:10]  # Changed 'df' to 'data'

# Extract relevant numerical columns
stats = ['attack', 'defense', 'speed']

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist([df_sample[stat] for stat in stats], bins=10, alpha=0.6, label=stats)

# Add title and legend
plt.title('Histogram of Attack, Defense, and Speed for First 10 Pokémon')
plt.xlabel('Stat Value')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```
